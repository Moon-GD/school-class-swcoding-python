```python
class Counter():
    def __init__(self, number = 0):
        if number >= 100 or number < 0:
            self.__number = 0
        else:
            self.__number = number
        
    def __str__(self):
        return f'C({self.__number})'
    
    def reset(self):
        self.__number = 0
        
    def inc(self):
        self.__number +=  1
        if self.__number >= 100:
            self.__number = 0
            
    def dec(self):
        self.__number -= 1
        if self.__number < 0:
            self.__number = 0
            
    def __add__(self, c1):
        new_number = self.__number + c1.__number
        if new_number >= 100:
            new_number = 0
            
        return Counter(new_number)
    
    def __sub__(self, c1):
        new_number = self.__number - c1.__number
        if new_number < 0 :
            new_number = 0
        
        return Counter(new_number)
```

```python
c1 = Counter(10)
c1.inc()
print('c1 = ', c1)
print()

c2 = Counter()
c2.inc()
c2.inc()
c2.dec()
print('c2 = ', c2)
c2.reset()
print('c2 = ', c2)
```

```python
c1 = Counter(10)
c2 = Counter(20)

c3 = c1 + c2
c4 = c1 - c2

print('c3 = ', c3)
print('c4 = ', c4)
```

```python
class BankAccount():
    def __init__(self, __name, account_num):
        self.__name = str(__name)
        self.__account_num = str(account_num)
        self.__balance = 0
        
    def __str__(self):
        return f'{self.__name}님의 계좌 {self.__account_num}의 잔고는 {self.__balance}원입니다.'
    
    def get_name():
        return self.__name
    
    def get_account_num():
        return self.__account_num
    
    def get_balance():
        return self.__balance
    
    def deposit(self, money):
        self.__balance += money
        print(f'{money}원이 입금되었습니다. 잔고는 {self.__balance}원입니다.')
        
    def withdraw(self, money):
        if self.__balance - money < 0:
            print(f'계좌 잔고는 {self.__balance}원으로 인출 요구 금액 {money}원보다 작습니다.')
        else:
            self.__balance -= money
```

```python
account1 = BankAccount('홍길동', '1234-0001')
print(account1)
account1.deposit(2000)
print(account1)
account1.withdraw(500)
print(account1)
account1.withdraw(5000)
```

```python
class Student():
    def __init__(self, name, student_id):
        self.__name = name
        self.__student_id = student_id
        
    def get_name(self):
        return self.__name
    
    def get_stuent_id(self):
        return self.__student_id
    
    def get_korean_quiz(self):
        return self.__korean_quiz
    
    def get_math_quiz(self):
        return self.__math_quiz
    
    def get_science_quiz(self):
        return self.__science_quiz
    
    def set_korean_quiz(self, score):
        self.__korean_quiz = score
    
    def set_math_quiz(self, score):
        self.__math_quiz = score
        
    def set_science_quiz(self, score):
        self.__science_quiz = score
        
    def get_total_score(self):
        self.__total_score = self.__korean_quiz + self.__math_quiz + self.__science_quiz
        return self.__total_score
        
    def get_avg_score(self):
        return self.__total_score / 3
    
    def __str__(self):
        return f'''이름 : {self.__name}, 학번 : {self.__student_id}
        국어성적 : {self.get_korean_quiz()}, 수학성적 : {self.get_math_quiz()}, 과학성적 : {self.get_science_quiz()}
        합계 : {self.get_total_score()}, 평균 : {self.get_avg_score()}'''
```

```python
# 객체 생성
name = input('학생의 이름을 입력하세요 : ')
student_id = input('학생의 학번을 입력하세요 : ')
student = Student(name, student_id)

# 점수 입력
korean_quiz = int(input('학생의 국어 성적을 입력하세요 : '))
math_quiz = int(input('학생의 수학 성적을 입력하세요 : '))
science_quiz = int(input('학생의 과학 성적을 입력하세요 : '))

# 점수 설정
student.set_korean_quiz(korean_quiz)
student.set_math_quiz(math_quiz)
student.set_science_quiz(science_quiz)

# 학생 정보 출력
print(student)
```

```python
class TV():
    # 클래스 변수 선언
    MIN_VOLUME = 0
    MAX_VOLUME = 20
    MIN_CHANNEL = 0
    MAX_CHANNEL = 200
    
    def __init__(self):
        self.__volume = 5
        self.__channel = 0
        self.__is_on = False
        
    def __str__(self):
        if self.__is_on:
            return f'TV 볼륨 : {self.__volume}, TV 채널 {self.__channel}'
        else:
            return 'TV가 꺼짐 상태입니다.'
        
    def toggle_power(self):
        self.__is_on = !self.__is_on
        
    def get_channel(self):
        return self.__channel
    
    def set_channel(self, choice):
        if choice > TV.MAX_CHANNEL or choice < TV.MIN_CHANNEL:
            print('채널 오류')
        else:
            self.__channel = choice
            
    def get_volume(self):
        return self.__volume
    
    def set_volume(self, choice):
        if choice > TV.MAX_VOLUME or choice < TV.MIN_VOLUME:
            print('볼륨 오류')
        else:
            self.__volume = choice
            
    def volume_up(self):
        self.__volume += 1
        if self.__volume > TV.MAX_VOLUME:
            self.__volume = TV.MAX_VOLUME
    
    def volume_down(self):
        self.__volume -= 1
        if self.__volume < TV.MIN_VOLUME:
            self.__volume = TV.MIN_VOLUME
    
    def channel_up(self):
        self.__channel += 1
        if self.__channel > TV.MAX_CHANNEL:
            self.__channel = TV.MIN_CHANNEL
    
    def channel_down(self):
        self.__channel -= 1
        if self.__channel < TV.MIN_CHANNEL:
            self.__channel = TV.MAX_CHANNEL
```

```python
my_tv = TV()
print(my_tv)
my_tv.toggle_power()
print(my_tv)
my_tv.set_channel(200)
print(my_tv)
my_tv.volume_up()
my_tv.channel_up()
print(my_tv)
```

```python
class Rectangle():
    def __init__(self, x, y, width, height):
        self.__x = x
        self.__y = y
        self.__width = width
        self.__height = height
        
    def __str__(self):
        return f'Rectangle : (x = {self.__x}, y = {self.__y}, h = {self.__height}), 면적 : {self.__width * self.__height}'
    
    def set_x(self, x):
        self.__x = x
        
    def set_x(self, y):
        self.__y = y
        
    def set_x(self, width):
        self.__width = width
        
    def set_x(self, height):
        self.__height = height
        
    def get_x(self):
        return self.__x
    
    def get_y(self):
        return self.__y
    
    def get_height(self):
        return self.__height
    
    def get_width(self):
        return self.__width
    
    def area(self):
        return self.__width * self.__height
    
    def overlaps(self, rec):
        if self.__x > rec.__x + rec.__width:
            return False
        if self.__x + self.__width < rec.__x:
            return False
        if self.__y > rec.__y + rec.__height:
            return False
        if self.__y + self.__height < rec.__y:
            return False
        return True
    
    def contain(self, rec):
        if self.__x > rec.__x or self.__x + self.__width < rec.__x + rec.__width:
            return False
        if self.__y < rec.__y or self.__y - self.__height > rec.__y - rec.__height:
            return False
        return True
```

```python
r1 = Rectangle(0, 0, 100, 100)
r2 = Rectangle(0, -10, 10, 10)
r3 = Rectangle(-100, 0, 120, 100)

print('r1 = ', r1)
print('r2 = ', r2)
print('r3 = ', r3)

print('r1 contains r2 : ', r1.contain(r2))  
print('r1 contains r3 : ', r1.contain(r3))
print('r1 overlaps r2 : ', r1.overlaps(r2))
print('r1 overlaps r3 : ', r1.overlaps(r3))
```

```python
class Circle():
    PI = 3.14
    
    def __init__(self, x, y, radius):
        self.__x = x
        self.__y = y
        self.__radius = radius
        
    def __str__(self):
        return f'Circle : (x = {self.__x}, y = {self.__y}, r = {self.__radius}), 면적 : {Circle.PI * self.__radius * self.__radius}'
    
    def set_x(self, new_x):
        self.__x = new_x
        
    def get_x(self):
        return self.__x
    
    def set_y(self, new_y):
        self.__y = new_y
        
    def get_y(self):
        return self.__y
    
    def set_radius(self, new_radius):
        self.__radius = new_radius
        
    def get_radius(self):
        return self.__radius
    
    def area():
        return self.__radius * self.__radius * Circle.PI
    
    def overlaps(self, c1):
        distance = ((self.__x - c1.__x) ** 2 + (self.__y - c1.__y) ** 2) ** 0.5
        if distance > self.__radius + c1.__radius:
            return False
        else:
            return True
    
    def contains(self, c1):
        distance = ((self.__x - c1.__x) ** 2 + (self.__y - c1.__y) ** 2) ** 0.5
        if distance < self.__radius:
            return True
        else:
            return False
```

```python
c1 = Circle(0, 0, 100)
c2 = Circle(0, -10, 10)
c3 = Circle(-100, 0 ,120)

print('c1 = ', c1)
print('c2 = ', c2)
print('c3 = ', c3)

print('c1 contains c2 : ', c1.contains(c2))
print('c1 contains c3 : ', c1.contains(c3))
print('c1 overlaps c2 : ', c1.overlaps(c2))
print('c1 overlaps c3 : ', c1.overlaps(c3))
```
